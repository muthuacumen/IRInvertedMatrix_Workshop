```python
import os
import requests
import feedparser

# Example: Download blog posts from a public RSS feed and save as .txt files
def fetch_and_save_blog_posts(rss_url, save_folder, max_posts=20):
    feed = feedparser.parse(rss_url)
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    count = 0
    for entry in feed.entries:
        if count >= max_posts:
            break
        title = entry.get('title', 'untitled')
        content = entry.get('summary', '')
        # Clean filename
        filename = f"blog_{count+1}.txt"
        filepath = os.path.join(save_folder, filename)
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(f"{title}\n\n{content}")
        count += 1
    print(f"Saved {count} blog posts to {save_folder}")

# Example RSS feed: Python Software Foundation blog
rss_url = 'https://pyfound.blogspot.com/feeds/posts/default?alt=rss'
fetch_and_save_blog_posts(rss_url, 'sample_docs', max_posts=20)
```

```python
# Example: Load text files from a folder
import os

def load_documents(folder_path):
    documents = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                documents.append(file.read())
    return documents

# Replace 'sample_docs/' with your actual folder
documents = load_documents('sample_docs/')
print(f"Loaded {len(documents)} documents.")
```

## ✂️ Step 2: Tokenizer


### 🗣 Instructor Talking Point:
> The tokenizer breaks raw text into a stream of words (tokens). This is the foundation for every later step in IR and NLP.

### 🔧 Your Task:
- Implement a basic tokenizer that splits text into lowercase words.
- Handle punctuation removal and basic non-alphanumeric filtering.


```python
import re

def tokenize(text):
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

# Test on one document
tokens = tokenize(documents[0])
print(tokens[:20])  # Preview first 20 tokens
```

## 🔁 Step 3: Normalization Pipeline (Stemming, Stop Word Removal, etc.)


### 🗣 Instructor Talking Point:
> Now we normalize tokens: convert to lowercase, remove stop words, apply stemming or affix stripping. This reduces redundancy and enhances search accuracy.

### 🔧 Your Task:
- Use `nltk` to remove stopwords and apply stemming.


```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def normalize_tokens(tokens):
    return [stemmer.stem(t) for t in tokens if t not in stop_words]

# Example: normalize one document
norm_tokens = normalize_tokens(tokens)
print(norm_tokens[:20])
```

## 🔍 Step 4: Inverted Index


### 🗣 Instructor Talking Point:
> We now map each normalized token to the list of document IDs in which it appears. This is the core structure that allows fast Boolean and phrase queries.

### 🔧 Your Task:
- Build the inverted index using a dictionary.
- Add code to support phrase queries using positional indexing.


```python
from collections import defaultdict

def build_inverted_index(documents):
    index = defaultdict(list)
    for doc_id, text in enumerate(documents):
        tokens = normalize_tokens(tokenize(text))
        seen = set()
        for token in tokens:
            if token not in seen:
                index[token].append(doc_id)
                seen.add(token)
    return dict(index)

inverted_index = build_inverted_index(documents)
print(dict(list(inverted_index.items())[:10]))  # Preview first 10 terms
```

## 🧪 Test: Phrase Queries


### 🗣 Instructor Talking Point:
> A phrase query requires the exact sequence of terms (e.g., "machine learning"). To support this, extend the inverted index to store positions, not just docIDs.

### 🔧 Your Task:

- Demonstrate that they return the correct documents.


```python
# Placeholder for phrase query implementation
# You may build a position-aware index or use string search within docs after normalization

# Example:
query1 = "Python software"
query2 = "bylaws"

def phrase_query(query, documents):
    results = []
    query_tokens = normalize_tokens(tokenize(query))
    for doc_id, doc in enumerate(documents):
        doc_tokens = normalize_tokens(tokenize(doc))
        # Search for the sequence in the token list
        for i in range(len(doc_tokens) - len(query_tokens) + 1):
            if doc_tokens[i:i+len(query_tokens)] == query_tokens:
                results.append(doc_id)
                break
    return results

# Run phrase queries on the loaded documents
results1 = phrase_query(query1, documents)
results2 = phrase_query(query2, documents)

print(f"Documents containing the phrase '{query1}': {results1}")
print(f"Documents containing the phrase '{query2}': {results2}")
```

```python
# Implement Positional Indexes: Store term positions for each document
from collections import defaultdict

def build_positional_index(documents):
    positional_index = defaultdict(lambda: defaultdict(list))
    for doc_id, text in enumerate(documents):
        tokens = normalize_tokens(tokenize(text))
        for pos, token in enumerate(tokens):
            positional_index[token][doc_id].append(pos)
    return positional_index

# Build positional index for the loaded documents
positional_index = build_positional_index(documents)

# Example: Show positions for a sample term
sample_term = 'python'  # Change to any term you want to inspect
print(f"Positions for term '{sample_term}':")
for doc_id, positions in positional_index[sample_term].items():
    print(f"  Document {doc_id}: {positions}")
```

### 📊 Comparison: Positional Index vs. Term Document Count Matrix

A **Positional Index** stores not only which documents a term appears in, but also the exact positions of each term within those documents. In contrast, a **Term Document Count Matrix** (TDCM) simply records the frequency of each term in each document, without any positional information.

| Feature                      | Positional Index                | Term Document Count Matrix (TDCM) |
|------------------------------|---------------------------------|-----------------------------------|
| Stores term positions        | Yes                             | No                                |
| Supports phrase/bigram search| Yes                             | No                                |
| Memory usage                 | Higher                          | Lower                             |
| Query speed for phrases      | Fast                            | Slow (requires scanning)          |
| Useful for                   | Phrase queries, proximity search | Keyword frequency analysis         |
| Implementation complexity    | Higher                          | Lower                             |

**Summary:**
- Use a positional index for advanced search features like phrase and proximity queries.
- Use a TDCM for simple keyword frequency analysis and ranking.

### 📐 Mathematical Explanation: Inverse Document Frequency (IDF)

The **Inverse Document Frequency (IDF)** is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The intuition is that terms that appear in many documents are less informative than those that appear in few.

The IDF for a term $t$ is defined as:

$$
\text{IDF}(t) = \log\left(\frac{N}{n_t}\right)
$$

Where:
- $N$ is the total number of documents in the corpus.
- $n_t$ is the number of documents containing the term $t$.

---

#### Example Calculation

Suppose:
- $N = 20$ (total documents)
- $n_{\text{softwar}} = 15$ (documents containing the term 'softwar')

Then:

$$
\text{IDF}(\text{softwar}) = \log\left(\frac{20}{15}\right) \approx 0.2877
$$

So, the IDF for term 'softwar' is $0.2877$.

A higher IDF value means the term is rare across the corpus, while a lower value means the term is common. IDF is often used in combination with Term Frequency (TF) to compute the TF-IDF score:

$$
\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)
$$

Where $\text{TF}(t, d)$ is the frequency of term $t$ in document $d$.

This weighting helps highlight terms that are both frequent in a document and rare across the corpus, improving search relevance and document ranking.

```python
# Calculate Term Frequency (TF) for the term 'softwar' in each document
# TF is simply the number of times the term appears in a document
import pandas as pd

term = 'softwar'  # Use the stemmed term as before

tf_values = []
for doc_id, doc in enumerate(documents):
    tokens = normalize_tokens(tokenize(doc))
    tf = tokens.count(term)
    tf_values.append(tf)
    print(f"TF for term '{term}' in Document {doc_id}: {tf}")

# Display as a DataFrame for clarity
TF_df = pd.DataFrame({'Document': list(range(len(documents))), 'TF': tf_values})
print(TF_df)
```

TF for term 'softwar' in Document 0:

$$

$$

TF-IDF for term 'softwar' in Document 0:

$$

$$

Where:

```python
# ============================================================
# Task completed by Muthu — Zipf's Law: Term Frequency Distribution
# ============================================================
# Zipf's Law says: a few words appear VERY often, and most words
# appear rarely. It's like social media followers — a handful of
# accounts have millions, but most people have a few hundred.

from collections import Counter

# Count how often each normalized term appears across the entire corpus
all_terms = [token for tokens in all_norm_tokens for token in tokens]
term_counts = Counter(all_terms)

# Sort by frequency (most common first)
sorted_terms = term_counts.most_common()
ranks = range(1, len(sorted_terms) + 1)
frequencies = [count for _, count in sorted_terms]

# --- Plot 1: Top 30 most frequent terms (bar chart) ---
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

top_n = 30
top_terms = sorted_terms[:top_n]
axes[0].barh([t for t, _ in top_terms][::-1], [c for _, c in top_terms][::-1], color='teal')
axes[0].set_xlabel('Frequency', fontsize=12)
axes[0].set_title(f'Top {top_n} Most Frequent Terms (After Normalization)', fontsize=13)

# --- Plot 2: Zipf's Law (log-log rank vs frequency) ---
axes[1].loglog(ranks, frequencies, marker='.', linestyle='none', color='darkorange', alpha=0.6)
axes[1].set_xlabel('Rank (log scale)', fontsize=12)
axes[1].set_ylabel('Frequency (log scale)', fontsize=12)
axes[1].set_title("Zipf's Law: Rank vs. Frequency (Log-Log)", fontsize=13)

plt.tight_layout()
plt.show()

print(f"\nMost common term: '{sorted_terms[0][0]}' appears {sorted_terms[0][1]} times")
print(f"Least common terms appear only {sorted_terms[-1][1]} time(s)")
print(f"Total unique terms: {len(sorted_terms)}")
```

**Key Insight — Zipf's Law in Action**

- The **left chart** shows the most common words in our corpus. Terms like "python", "p", and "board" dominate because our data comes from Python Software Foundation blog posts.
- The **right chart** plots every term's rank against its frequency on a log-log scale. If the dots form a roughly straight line sloping downward, our corpus follows **Zipf's Law** — a well-known pattern in natural language where a small number of words do most of the heavy lifting.
- **Why does this matter for search?** The most frequent terms (like "python") appear in almost every document, so they are *not* very useful for distinguishing one document from another. This is exactly why **IDF** (Inverse Document Frequency) is important — it gives higher weight to rare, distinctive terms.

> Analogy: If every student in your class owns a backpack, "owns a backpack" tells you nothing about who is who. But "plays the tuba" — that narrows it down fast. IDF works the same way.

```python
# ============================================================
# Task completed by Muthu — TF-IDF Heatmap: Top Terms x Documents
# ============================================================
# A heatmap is like a color-coded grade sheet: rows are terms,
# columns are documents, and brighter colors mean higher TF-IDF
# scores — i.e., that term is especially important in that document.

# Step 1: Pick the top 15 terms by IDF (most distinctive terms)
top_idf_terms = sorted(idf_scores.items(), key=lambda x: x[1], reverse=True)[:15]
top_terms_list = [term for term, _ in top_idf_terms]

# Step 2: Build a TF-IDF matrix (rows=terms, columns=documents)
tfidf_matrix = []
for term in top_terms_list:
    row = []
    for doc_id in range(len(documents)):
        # TF = number of times this term appears in this document
        tf = all_norm_tokens[doc_id].count(term)
        # IDF = already computed
        idf = idf_scores.get(term, 0.0)
        row.append(tf * idf)
    tfidf_matrix.append(row)

# Step 3: Create a DataFrame and plot the heatmap
tfidf_df = pd.DataFrame(
    tfidf_matrix,
    index=top_terms_list,
    columns=[f'Doc {i}' for i in range(len(documents))]
)

fig, ax = plt.subplots(figsize=(16, 7))
sns.heatmap(
    tfidf_df,
    annot=True,          # show the numeric values inside each cell
    fmt='.2f',           # round to 2 decimal places
    cmap='YlOrRd',       # yellow-to-red color gradient
    linewidths=0.5,
    ax=ax
)
ax.set_title('TF-IDF Heatmap: Top 15 Distinctive Terms Across Documents', fontsize=14)
ax.set_xlabel('Document', fontsize=12)
ax.set_ylabel('Term (stemmed)', fontsize=12)
plt.tight_layout()
plt.show()

print("Brighter/darker red cells = higher TF-IDF = term is especially relevant to that document.")
```

**What do we see here?**

- Most cells are **pale yellow (near zero)** — this means most distinctive terms only appear in one or two documents. That's exactly what makes them distinctive!
- The **bright red hotspots** tell us which document "owns" a particular topic. For example, if the term "bylaw" lights up only in Documents 0 and 14, those are likely the documents about PSF governance changes.
- Terms that are uniformly zero across most documents have **high IDF** — they are rare. When they *do* appear, they strongly signal what that document is about.

> Think of it like a class seating chart with colored stickers: if a sticker only appears on one desk, you immediately know whose it is. That's high TF-IDF in action.

```python
# ============================================================
# Task completed by Muthu — Boolean vs. TF-IDF Ranked Search
# ============================================================
# Boolean search just says "yes/no, this document matches."
# TF-IDF ranked search says "this document matches, and HERE'S
# HOW RELEVANT it is." Let's compare them side by side.

def boolean_search(query, inverted_index):
    """Return document IDs that contain ALL query terms (AND logic)."""
    query_terms = normalize_tokens(tokenize(query))
    if not query_terms:
        return []
    # Start with docs matching the first term, then intersect
    result_set = set(inverted_index.get(query_terms[0], []))
    for term in query_terms[1:]:
        result_set &= set(inverted_index.get(term, []))
    return sorted(result_set)

def ranked_search(query, documents, idf_scores):
    """Return documents ranked by total TF-IDF score for the query."""
    query_terms = normalize_tokens(tokenize(query))
    scores = {}
    for doc_id in range(len(documents)):
        doc_tokens = all_norm_tokens[doc_id]
        total_score = 0.0
        for term in query_terms:
            tf = doc_tokens.count(term)
            idf = idf_scores.get(term, 0.0)
            total_score += tf * idf
        if total_score > 0:
            scores[doc_id] = round(total_score, 4)
    # Sort by score descending
    return dict(sorted(scores.items(), key=lambda x: x[1], reverse=True))

# --- Run both search methods on the same query ---
search_query = "python community"
print(f"Search query: '{search_query}'\n")

boolean_results = boolean_search(search_query, inverted_index)
ranked_results = ranked_search(search_query, documents, idf_scores)

print(f"Boolean Search (AND): {len(boolean_results)} documents match → {boolean_results}")
print(f"Ranked Search (TF-IDF): {len(ranked_results)} documents, sorted by relevance:")
for doc_id, score in ranked_results.items():
    print(f"  Doc {doc_id}: TF-IDF score = {score}")

# --- Side-by-side bar chart ---
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Left: Boolean results (binary — match or no match)
boolean_flags = [1 if i in boolean_results else 0 for i in range(len(documents))]
colors_bool = ['seagreen' if f == 1 else 'lightgray' for f in boolean_flags]
axes[0].bar(range(len(documents)), boolean_flags, color=colors_bool)
axes[0].set_xlabel('Document ID', fontsize=12)
axes[0].set_ylabel('Match (1=Yes, 0=No)', fontsize=12)
axes[0].set_title(f'Boolean AND Search: "{search_query}"', fontsize=13)
axes[0].set_xticks(range(len(documents)))
axes[0].set_yticks([0, 1])

# Right: Ranked results (continuous scores)
ranked_scores_all = [ranked_results.get(i, 0.0) for i in range(len(documents))]
colors_rank = ['darkorange' if s > 0 else 'lightgray' for s in ranked_scores_all]
axes[1].bar(range(len(documents)), ranked_scores_all, color=colors_rank)
axes[1].set_xlabel('Document ID', fontsize=12)
axes[1].set_ylabel('TF-IDF Relevance Score', fontsize=12)
axes[1].set_title(f'TF-IDF Ranked Search: "{search_query}"', fontsize=13)
axes[1].set_xticks(range(len(documents)))

plt.tight_layout()
plt.show()
```

**What do we see here?**

- **Boolean search (left)** is all-or-nothing: a document either contains *all* query terms or it doesn't. Every matching document looks the same — there's no ranking.
- **TF-IDF ranked search (right)** shows a gradient: some documents are *much more* relevant than others. The tallest orange bars are the best matches — they mention the query terms frequently *and* those terms are distinctive.
- Notice how ranked search can surface the same documents as Boolean, but it also tells you **which one to read first**. That's the real power of TF-IDF.

> Analogy: Boolean search is like asking "who in this room has read Harry Potter?" — everyone raises their hand. Ranked search is like asking "who can quote the most passages?" — now you know who the real fans are.

```python
# Query processor: given a query string and an inverted index,
# return a list of all document IDs that contain any of the query terms (Boolean OR).
def query_processor(query, inverted_index):
    results = []
    for term in query.split():
        if term in inverted_index:
            results.extend(inverted_index[term])
    return list(set(results))  # return unique doc IDs
```
